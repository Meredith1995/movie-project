Create seperate table with the each genre as a column 

```python
df3 = pd.DataFrame(columns=['tconst', 'genres']) #create the new dataframe using tconst and genres
d_list = [] # create a list to append with the new column names

for index, row in IMDB_titlesB.iterrows():
    for value in str(row['genres']).split(','):  #iterate through the original dataframe genres and split each string at ',' and make those the values
        d_list.append({'tconst':row['tconst'], 
                       'value':value})  #append the list with a dictionary of the new values
df3 = df3.append(d_list, ignore_index=True) #append the dataframe with the list of each tconst and genre
df3 = df3.groupby('tconst')['value'].value_counts() #groupby tconst and value(genres) and collect the value counts
df3 = df3.unstack(level=-1).fillna(0) #unstack the data from so genres run across the top and fill null values with 0
df3
```

```python
IMDB_titlesB.info() #checking values on the original data frame 
```

```python
IMDB_titlesB.isna().sum() #seeing what data is missing
```

```python
IMDB_titlesB.drop(['genres'], axis = 1, inplace = True) #drop old genre column
```

```python
IMDB_titlesB.set_index(['tconst'], inplace = True) #set the index to tconst to easily join both tables
```

```python
IMDB_titlesB.head() #check that the formating is correct
```

```python
df2 = IMDB_titlesB.join(df3, how = 'outer') #join the genre table and the original table with all the values on the their index
df2 
```

```python
Movies.drop(['Unnamed: 0'], axis = 1, inplace = True) #drop unnecessary column
```

```python
Movies['genre_ids'] = Movies['genre_ids'].str.strip('[]') #take away the brakets on the genre_ids 
```

```python
df5 = pd.DataFrame(columns=['id', 'genre_ids']) #repeat same steps from previous dataframe to split the genres into columns
c_list = []

for index, row in Movies.iterrows():
    for value in str(row['genre_ids']).split(','):
        c_list.append({'id':row['id'], 
                       'value':value})
df5 = df5.append(c_list, ignore_index=True)
df5 = df5.groupby('id')['value'].value_counts()
df5 = df5.unstack(level=-1).fillna(0)
df5
```

```python
#replace the number column names with the associated genre
df5.rename(columns= {'28':"Action",'12':"Adventure",'16':"Animation",'35':"Comedy",'80':"Crime",'99':"Documentary",'18':"Drama",'10751':"Family",'14':"Fantasy",'36':"History",'27':"Horror",'10402':"Music",'9648':"Mystery",'10749':"Romance",'878':"Science Fiction",'10770':"TV Movie",'53':"Thriller",'10752':"War",'37':"Western"}, inplace = True)
```
