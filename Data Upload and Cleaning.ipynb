```python
df3 = pd.DataFrame(columns=['tconst', 'genres'])
d_list = []

for index, row in IMDB_titlesB.iterrows():
    for value in str(row['genres']).split(','):
        d_list.append({'tconst':row['tconst'], 
                       'value':value})
df3 = df3.append(d_list, ignore_index=True)
df3 = df3.groupby('tconst')['value'].value_counts()
df3 = df3.unstack(level=-1).fillna(0)
df3
```
